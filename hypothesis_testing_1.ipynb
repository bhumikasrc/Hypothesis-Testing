```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

#### **Hypothesis Testing**

##### **Our claim:** MSDS students, on average, study more than 25 hours a day

**H0:** Avg. MSDS Student Study Time **=** 25 hrs

**H1:** Avg. MSDS Student Study Time **>** 25 hrs

```python
# Study ours sample of 100 MSDS students
study_hrs = np.random.normal(28,1,100)
study_hrs
```

```python
alpha = 0.05
n = len(study_hrs)
t_critical = stats.t.isf(alpha/2,df=n-1)
t_critical
# If t_stat is greater that t_critical, then we reject null hypothesis
```

```python
t_stat = (study_hrs.mean() - 25)/study_hrs.std()
t_stat
```

```python
sns.kdeplot(study_hrs)
plt.axvline(25,c='r')
```
