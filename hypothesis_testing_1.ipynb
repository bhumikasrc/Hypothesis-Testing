```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

#### **Hypothesis Testing**

##### **Our claim:** MSDS students, on average, study more than 25 hours a day

**H0:** Avg. MSDS Student Study Time **=** 25 hrs

**H1:** Avg. MSDS Student Study Time **>** 25 hrs

```python
# Study hours sample of 100 MSDS students
study_hrs = np.random.normal(28,1.7,100).round(2)
print(study_hrs)
```

```python
alpha = 0.05
n = len(study_hrs)
# Getting the t-alpha that represents a p-value of 5%
t_critical = stats.t.isf(alpha,df=n-1)
print('t-ctitical:',t_critical)
# If t_stat is greater that t_critical, then we reject null hypothesis
```

```python
# Calculating the t-statistic to compare where it places on the t distribution compared to the t-critical
t_stat = (study_hrs.mean() - 25)/study_hrs.std()
print('t-stat:',t_stat)
```

```python
print(t_stat,'>',t_critical)
```

```python
plt.figure(figsize=(10,7))
plt.subplot(2,1,1)
sns.kdeplot(study_hrs)
plt.axvline(25,c='r')
plt.title('Distribution of MSDS Student Study Hours')
plt.subplot(2,1,2)
sns.boxplot(study_hrs,orient='h')
plt.axvline(25,c='r',clip_on=False)
plt.xlabel('Study Hours')
plt.show()
```

```python
# Calculating the p-value
p_val = stats.t.sf(t_stat,df=n-1)
print(p_val,'< 0.05')
```
