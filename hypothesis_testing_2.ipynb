# Hypothesis Test: Is there a significant difference in mean height for Males and Females

### $$H_0: μ_f = μ_m$$
### $$H_a: μ_f \neq μ_m$$

### Calculate the test statistic and determine p-value

```python
# Step 1: Data Preparation
# Select a subset of the data for analysis
height_f = obesity[obesity['gender'] == 'Female']['height'][:100]
height_m = obesity[obesity['gender'] == 'Male']['height'][:100]
```

```python
# Step 2: Perform the Hypothesis Test
# Calculate the test statistic and p-value
t_statistic_height, p_value_height = stats.ttest_ind(height_f, height_m)
```

```python
# Step 3: Interpret the Results
# Check if the p-value is less than the significance level (e.g., 0.05)
if p_value_height < 0.05:
    print("Reject the null hypothesis")
```

#### Since p-values for the test are below the alpha level of 0.05 we reject the null hypothesis. There is a significant difference in the mean height between females and males 

# Mean difference of Height for the Population

```python
# Step 4: Data Visualization
# Create a boxplot to visualize the difference in mean height
sns.boxplot(x='gender', y='height', data=obesity, palette=["pink", "lightblue"])
plt.title("Gender vs Height")
plt.show()
```

#### We can also see from the boxplot above that there is indeed a difference in the mean height between males and females
